```python
import rawpy
import imageio
import os

# Carpeta raíz donde buscar los archivos CR2
input_folder = 'C:/Users/User/Desktop/FotosAntartida/antartida_copia'  # Cambia esto por tu carpeta raíz

for root, dirs, files in os.walk(input_folder):
    for filename in files:
        if filename.lower().endswith('.cr2'):
            input_path = os.path.join(root, filename)
            output_path = os.path.splitext(input_path)[0] + '.jpg'

            with rawpy.imread(input_path) as raw:
                rgb = raw.postprocess(use_camera_wb=True)
            imageio.imwrite(output_path, rgb)

            print(f'Convertido: {input_path} → {output_path}')
```
